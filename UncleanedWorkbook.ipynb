```python
merged2=pd.read_csv("UsedData/Merged_ratings_movies_revenue.csv")

merged2['ROI']=np.array(merged2.profit)/np.array(merged2.budget)
merged2.sort_values('ROI',ascending=False,inplace=True)
merged2=merged2.loc[merged2['revenue']>0,['title','revenue','profit','ROI','Action','budget','title']].loc[(merged2['budget']!=0),:].loc[merged2['Action']==1,:]
merged2=merged2.iloc[1:]
merged2
#merged2.ROI.describe()
```

```python
merged=pd.read_csv("UsedData/Merged_ratings_movies_revenue.csv")
def getGenreSet(dataset,genre):
    tempSubset = dataset
    tempSubset = tempSubset.loc[tempSubset['revenue']>0,:].loc[(tempSubset['budget']>100),:].loc[tempSubset[genre]==1,:].loc[(tempSubset['title']!="The Tiger: An Old Hunter's Tale")]
    tempSubset['ROI']=np.array(tempSubset.profit)/np.array(tempSubset.budget)
    tempSubset.sort_values('ROI',ascending=False,inplace=True)
    return tempSubset
    
    
    
```

```python
#logbins = np.logspace(merged2.ROI.min(),merged2.ROI.max(),base=np.e)
figHist = plt.figure(figsize=(30,20))
plt.subplot(2,2,1)
n, bins, patches=plt.hist(merged2.ROI,int(merged2.ROI.describe()[0]/10),(merged2.ROI.min(),merged2.ROI.mean()+merged2.ROI.std()+.5),color='pink')
y = np.linspace(0,max(n),50)

xmean = np.full((50,1),merged2.ROI.mean())
xStdLow = np.full((50,1),merged2.ROI.mean()-merged2.ROI.std())
xStdHigh = np.full((50,1),merged2.ROI.mean()+merged2.ROI.std())

plt.plot(xmean,y, label="Mean")
#plt.plot(xStdLow,y,ls='--',color='lightblue',label="Upper, Lower Stdev")
plt.plot(xStdHigh,y,ls='--',color='lightblue',label="Upper Stdev")
plt.title('Frequency of ROI in Action movies (N='+ str(len(merged2.ROI))+')')
plt.xlabel('Return on Investment')
plt.ylabel('Frequency (# of movies)')
plt.legend(loc=1)
ticks = plt.xticks(np.arange(int(min(bins)), int(max(bins))+1, 1.0))
#Second subplot
plt.subplot(2,2,2)
tempset = getGenreSet(merged,'Adventure')
n,bins,patches=plt.hist(tempset.ROI,int(tempset.ROI.describe()[0]/10),(tempset.ROI.min(),(tempset.ROI.mean()+tempset.ROI.std()+.5)),color='green')
y = np.linspace(0,max(n),50)
plt.plot(np.full((50,1),tempset.ROI.mean()),np.linspace(0,max(n),50),label='Mean')
ticks = plt.xticks(np.arange(int(min(bins)), int(max(bins))+1, 1.0))
plt.plot(np.full((50,1),tempset.ROI.mean()+tempset.ROI.std()),y,ls='--',color='lightblue',label="Upper Stdev")
plt.title('Frequency of ROI in Adventure movies (N='+ str(len(tempset.ROI))+')')
plt.xlabel('Return on Investment')
plt.ylabel('Frequency (# of movies)')
plt.legend(loc=1)
#Third subplot
plt.subplot(2,2,3)
tempset = getGenreSet(merged,'Drama')
n,bins,patches=plt.hist(tempset.ROI,int(tempset.ROI.describe()[0]/10),(tempset.ROI.min(),(tempset.ROI.mean()+tempset.ROI.std()+.5)),color='red')
y = np.linspace(0,max(n),50)
plt.plot(np.full((50,1),tempset.ROI.mean()),np.linspace(0,max(n),50),label='Mean')
ticks = plt.xticks(np.arange(int(min(bins)), int(max(bins))+1, 1.0))
plt.plot(np.full((50,1),tempset.ROI.mean()+tempset.ROI.std()),y,ls='--',color='lightblue',label="Upper Stdev")
plt.title('Frequency of ROI in Drama movies (N='+ str(len(tempset.ROI))+')')
plt.xlabel('Return on Investment')
plt.ylabel('Frequency (# of movies)')
plt.legend(loc=1)
#Fourth subplot
plt.subplot(2,2,4)
tempset = getGenreSet(merged,'Animation')
n,bins,patches=plt.hist(tempset.ROI,int(round(tempset.ROI.describe()[0]/10)),(tempset.ROI.min(),(tempset.ROI.mean()+tempset.ROI.std()+.5)),color='orange')
y = np.linspace(0,max(n),50)
plt.plot(np.full((50,1),tempset.ROI.mean()),np.linspace(0,max(n),50),label='Mean')
ticks = plt.xticks(np.arange(int(min(bins)), int(max(bins))+1, 1.0))
plt.plot(np.full((50,1),tempset.ROI.mean()+tempset.ROI.std()),y,ls='--',color='lightblue',label="Upper Stdev")
plt.title('Frequency of ROI in Animation movies (N='+ str(len(tempset.ROI))+')')
plt.xlabel('Return on Investment')
plt.ylabel('Frequency (# of movies)')
plt.legend(loc=1)
plt.show()
figHist.savefig('mwangPlotImages/genreHistogram.png')
```
